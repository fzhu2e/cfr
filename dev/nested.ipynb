```python
import cfr
import matplotlib.pyplot as plt
cfr.__version__
```

```python
pdb = cfr.ProxyDatabase().fetch('PAGES2kv2')
```

```python
fig, axs = plt.subplots(nrows=5, ncols=2, figsize=[10,10])
axs = axs.flatten()
start = 1000
for k in range(9):
    # find a nest
    century = [start+k*100+1, start+(k+1)*100]
    pid = pdb.nest_indices(time_period=century)  
    # filter
    comp = pdb.filter(by='pid', keys=pid).make_composite(bin_width=5,n_bootstraps=10)
    axs[k] = comp.plot_composite(ax=axs[k])
    axs[k]['var'].set_title(f'{century}') 
    
fig.tight_layout()
```
